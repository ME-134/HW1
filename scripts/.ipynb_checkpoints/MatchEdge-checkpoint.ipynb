```python
def find_first(line):
    for i in range(len(line)):
        if line[i] > 0:
            return i
    return np.nan

def find_horizontal_offset(img, c1, c2):
    contour_map = np.zeros((img.shape[0], img.shape[1]))
#     c1 = c1[c1[:, 0] > np.quantile(c1[:, 0], 0.5)] # Left
#     c2 = c2[c2[:, 0] < np.quantile(c2[:, 0], 0.5)] # Right
    
    cv2.drawContours(contour_map, c1.reshape(-1, 1, 2), -1, 255, 3)
    cv2.drawContours(contour_map, c2.reshape(-1, 1, 2), -1, 255, 3)
    
    mid = int(np.mean(np.concatenate([c1[:, 0], c2[:, 0]])))

    low = max(c1.min(axis = 0)[1], c2.min(axis = 0)[1])
    hi = min(c1.max(axis = 0)[1], c2.max(axis = 0)[1])
    spaces = np.linspace(low, hi, 30).astype(int)

    shifts = []
    plt.imshow(contour_map)
    for i in range(len(spaces)):
        mid_offset = find_first(contour_map[spaces[i]][mid:])
        reverse = contour_map[spaces[i]][:mid][::-1]
        start = mid - find_first(reverse)
        plt.plot([start, mid + mid_offset], [spaces[i], spaces[i]])
        shifts.append(mid + mid_offset - start)
    
    return np.nanmedian(shifts[1:-1])

find_horizontal_offset(img, c1, c2)
```

```python
def find_vertical_offset(img, c1, c2):
    contour_map = np.zeros((img.shape[0], img.shape[1]))
#     c1 = c1[c1[:, 1] < np.quantile(c1[:, 1], 0.5)] # Bottom 
#     c2 = c2[c2[:, 1] > np.quantile(c2[:, 1], 0.5)] # Top
    cv2.drawContours(contour_map, c1.reshape(-1, 1, 2), -1, 255, 3)
    cv2.drawContours(contour_map, c2.reshape(-1, 1, 2), -1, 255, 3)
    mid = int(np.mean(np.concatenate([c1[:, 1], c2[:, 1]])))

    low = max(c1.min(axis = 0)[0], c2.min(axis = 0)[0])
    hi = min(c1.max(axis = 0)[0], c2.max(axis = 0)[0])
    spaces = np.linspace(low, hi, 30).astype(int)
    shifts = []
    
    plt.imshow(contour_map)
    plt.plot([0, 270], [mid, mid])
    for i in range(len(spaces)):
        print(mid, contour_map.shape)
        mid_offset = find_first(contour_map[:, spaces[i]][mid:])
        reverse = contour_map[:, spaces[i]][:mid][::-1]
        start = mid - find_first(reverse)
        plt.plot([spaces[i], spaces[i]], [start, mid + mid_offset])
        shifts.append(mid + mid_offset - start)
    
    return np.nanmedian(shifts[1:-1])

find_vertical_offset(img, c1, c2)
```
